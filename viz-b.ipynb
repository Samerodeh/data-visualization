## Load Dataset

```python
planets = sns.load_dataset("planets")
planets.head()
```

## plot planets Dataset

### Count the (mass)

```python
sns.countplot(data=planets, x="mass")
```

### planets (method) based on (distance)

```python
sns.catplot(data=planets, x="method", y="distance")
```

### relplot =(orbital_period) based on (year)

```python
sns.relplot(data=planets, x="orbital_period",y="year")
```

### boxen the (number)

```python
sns.boxenplot(data=planets, x="number") 
```

### barplot (mass) and (distance)

```python
sns.barplot(data=planets, x="mass", y="distance") 
```
