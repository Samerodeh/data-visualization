## Load Dataset

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

## plot titanic Dataset

### Count the Pclass

```python
sns.countplot(data=titanic, x="pclass")
```

### strip the (age) and (sex)

```python
sns.stripplot(data=titanic, x="age", y="sex") 
```

### box plot (fare)

```python
sns.boxplot(data=titanic, x="fare")
```

### point the (alone) and (class)

```python
sns.pointplot(data=titanic, x="alone", y="class")
```

### violin (survived)

```python
sns.violinplot(data=titanic, x="survived")
```
